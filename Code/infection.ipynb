```python
time_axis = []
for i in range(60):
     time_axis.append(i)
```

```python
inf_hosts_axis = []
sum_num = 0
for i in range(len(num_inf_hosts)):
    sum_num = sum_num + num_inf_hosts[i]
    
    inf_hosts_axis.append(sum_num)
    
```

```python
#import matplotlib.pyplot as plt

plt.plot( time_axis, inf_hosts_axis, color='blue', marker='o')
plt.title('unauthorized Hosts detected (9-10)am', fontsize=18, color='black')
plt.xlabel('time in minutes', fontsize=14, color='black')
plt.ylabel('unauthorized hosts', fontsize=14, color='black')
plt.grid(True)
plt.show()
```
